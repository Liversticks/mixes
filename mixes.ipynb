```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G = nx.Graph()
```

```python
spirits = [
    "Angostura Bitters",
    "Bourbon",
    "Calvados",
    "Campari",
    "Cider",
    "Cognac",
    "Gin",
    "Green Chartreuse",
    "Kahlua", # or other coffee liqueur
    "Luxardo", # or other maraschino liqueur
    "Mead", # honey wine, ~15-20% ABV
    "White Rum",
    "Tequila",
    "Triple Sec", # orange curacao, includes variations such as Grand Marnier and Cointreau
    "Sweet Vermouth", # can sub with brandy umeshu
    "Vodka"
    ]

mixers = [
    "Cream",
    "Lemon Juice",
    "Lime Juice",
    "Simple Syrup",
    "Yakult" # or other cultured liquid yogurt
]

G.add_nodes_from(spirits)
G.add_nodes_from(mixers)
```

```python
def generate_edges_from_ingredients(*args):
    for arg1 in args:
        for arg2 in args:
            if arg1 != arg2:
                G.add_edge(arg1, arg2)
```

```python
# define recipes

# Alexander
generate_edges_from_ingredients("Cognac", "Cream", "Kahlua")
generate_edges_from_ingredients("Cognac", "Triple Sec", "Cream", "Kahlua")
generate_edges_from_ingredients("Gin", "Cream", "Kahlua")

# Boulevardier
generate_edges_from_ingredients("Bourbon", "Sweet Vermouth", "Campari")

# Corpse Reviver #1
generate_edges_from_ingredients("Cognac", "Calvados", "Sweet Vermouth")

# Last Word
generate_edges_from_ingredients("Gin", "Green Chartreuse", "Luxardo", "Lime Juice")

# Lemon Drop
generate_edges_from_ingredients("Vodka", "Triple Sec", "Lemon Juice")

# Long Island Iced Tea
generate_edges_from_ingredients("Gin", "Vodka", "Tequila", "White Rum", "Triple Sec", "Simple Syrup")

# Manhattan
generate_edges_from_ingredients("Bourbon", "Sweet Vermouth", "Angostura Bitters")

# Man of War
generate_edges_from_ingredients("Bourbon", "Triple Sec", "Sweet Vermouth", "Lemon Juice")

# Margarita
generate_edges_from_ingredients("Tequila", "Lime Juice", "Triple Sec")

# Negroni
generate_edges_from_ingredients("Gin", "Sweet Vermouth", "Campari")

# Sidecar
generate_edges_from_ingredients("Cognac", "Triple Sec", "Lemon Juice")

# Experimental
generate_edges_from_ingredients("Cider", "Sweet Vermouth")
generate_edges_from_ingredients("Gin", "Calvados", "Sweet Vermouth", "Cider")
generate_edges_from_ingredients("Mead", "Cognac", "Sweet Vermouth")
generate_edges_from_ingredients("Yakult", "Gin", "Triple Sec", "Angostura Bitters") # originally tested with campari over bitters, but too strong
```

```python
nx.draw_kamada_kawai(G, with_labels=True, font_weight='bold')
plt.show()
```
